```python
new_cols = ['name', 'description', 'purl', 'image_url', 'manufacturer','cell','wattage_per_panel','weight','commercial_residential','wattage','cell_technology','height']
data = data[new_cols]
```

```python
data = data.astype(str)
print(data.dtypes)
```

```python
product_strings = []
for index, row in data.iterrows():
    product_name = row['name']
    product_url = row['purl']
    #image_url = row['image_url']
    product_string = f'[{product_name}]("{product_url}")'
    product_strings.append(product_string)

# Add the product strings as a new column in the DataFrame
data['product_string'] = product_strings
```
